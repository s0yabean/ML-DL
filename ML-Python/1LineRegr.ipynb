# Machine Learning Based on Python

### Linear Regression
#### Generate Data using the function y = x/2 + cos(x) + Gaussian Noise (0, 1)

```python
import numpy as np
from sklearn import linear_model, datasets, tree
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
no_samples = 100
x = np.linspace(-np.pi, np.pi, no_samples)
y = 0.5*x + np.cos(x) + np.random.random(x.shape)
plt.scatter(x,y,color='red') #Plot y-vs-x in dots
plt.xlabel('Input-x (feature)')
plt.ylabel('Output-y (Target)')
plt.title('Fig 1: Data for linear regression')
plt.show()
```

#### Randonly split the Dataset into Training (70%), Validation (15%), and Testing (15%)

```python
random_indices = np.random.permutation(no_samples)
#Training set
x_train = x[random_indices[:70]]
y_train = y[random_indices[:70]]
#Validation set
x_val = x[random_indices[70:85]]
y_val = y[random_indices[70:85]]
#Test set
x_test = x[random_indices[85:]]
y_test = y[random_indices[85:]]
```

```python
print('Train: ',len(x_train))
print('Valid: ',len(x_val))
print('Test: ',len(x_test))
```

### Fit a line to the Data

```python
model = linear_model.LinearRegression()

#Reshpae the arrays into column matrices
x_train_for_line_fitting = np.matrix(x_train.reshape(len(x_train),1))
y_train_for_line_fitting = np.matrix(y_train.reshape(len(y_train),1))

#Fit the line to the training data
model.fit(x_train_for_line_fitting, y_train_for_line_fitting)

#Plot the line
plt.scatter(x_train, y_train, color='red')
plt.plot(x.reshape((len(x),1)),model.predict(x.reshape((len(x),1))),color='black', lw=3)
plt.xlabel('Input-x (Feature)')
plt.ylabel('Output-y (Target values)')
plt.title('Fig 2: Fitting the training data with a straight Line')
plt.show()
```

### Evaluate the Model
#### Evaluate the Model in terms of Mean Square Error (MSE) on the Validation and Test Datasets.

```python
Validation_error = np.mean( (y_val - model.predict(x_val.reshape(len(x_val),1)))**2 )
Test_error = np.mean( (y_test - model.predict(x_test.reshape(len(x_test),1)))**2 )

print ('Validation MSE: ', Validation_error, '\nTest MSE: ', Test_error)
```
